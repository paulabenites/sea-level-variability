```python
import scrapy

from scrapy.linkextractors.lxmlhtml import LxmlLinkExtractor
from scrapy.spiders import Rule, CrawlSpider
from scrapy.http import Request
from scrapy.exceptions import CloseSpider



class YearItem(scrapy.Item):
    enero = scrapy.Field()
    febrero = scrapy.Field()
    marzo = scrapy.Field()
    abril = scrapy.Field()
    mayo = scrapy.Field()
    junio = scrapy.Field()
    julio = scrapy.Field()
    agosto = scrapy.Field()
    septiembre = scrapy.Field()
    octubre = scrapy.Field()
    noviembre = scrapy.Field()
    diciembre = scrapy.Field()
    anio = scrapy.Field()

class ClimaSpider(scrapy.Spider):
    """
    This CrawlSpider will get information about jobs. It will navigate through the links of the page
    """
    name = 'clima'
    start_urls = ['https://origin.cpc.ncep.noaa.gov/products/analysis_monitoring/ensostuff/ONI_v5.php']
    
    
    
    
    
    def parse(self, response):
        """
        This function takes the information from each page and return the items in the catalog.
        """
        self.logger.info('Now we are crawling: %s', response.url)
        print(response.xpath('//html'))
        i=0
        for row in response.xpath('//table[@align="center"][@border=1]//tr'):
          i+=1
          if (i-1) % 11 == 0:
            pass
          else:
            year = YearItem()
            year['anio']       = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[1]//strong/text()').get()
            year['enero']      = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[2]//font/text()').get()
            year['febrero']    = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[3]//font/text()').get()
            year['marzo']      = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[4]//font/text()').get()
            year['abril']      = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[5]//font/text()').get()
            year['mayo']       = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[6]//font/text()').get()
            year['junio']      = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[7]//font/text()').get()
            year['julio']      = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[8]//font/text()').get()
            year['agosto']     = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[9]//font/text()').get()
            year['septiembre'] = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[10]//font/text()').get()
            year['octubre']    = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[11]//font/text()').get()
            year['noviembre']  = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[12]//font/text()').get()
            year['diciembre']  = response.xpath('//table[@align="center"][@border=1]//tr['+str(i)+']/td[13]//font/text()').get()
            
            yield year
    


```

```python
from scrapy.crawler import CrawlerProcess

process = CrawlerProcess(settings={
           'FEEDS': {
               'clima.csv' : {
                   'format' : 'csv',
                   'fields': ['anio', 'enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre']
               }
           }
          }
    )

process.crawl( ClimaSpider )
process.start()
```
